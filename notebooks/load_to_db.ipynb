```python
from sqlalchemy import create_engine, text
import pymysql
from urllib.parse import quote_plus

user = "root"
raw_password = "Abhi@2026"
password = quote_plus(raw_password)

host = "localhost"
port = 3306
db = "stock_analysis_db"

url = f"mysql+pymysql://{user}:{password}@{host}:{port}/{db}"
print("URL:", url)

engine = create_engine(url)

with engine.connect() as conn:
    result = conn.execute(text("SELECT DATABASE();"))
    print("Connected to DB:", list(result))
```

```python
import pandas as pd

daily_df = pd.read_csv("../data_processed/all_stocks_daily.csv")

daily_df.to_sql(
    "stocks_daily",
    engine,
    if_exists="replace",  
    index=False
)
```

```python
yearly_sector_df = pd.read_csv("../data_processed/yearly_return_with_sector.csv")
yearly_sector_df.to_sql("stock_yearly_metrics", engine, if_exists="replace", index=False)

monthly_df = pd.read_csv("../data_processed/monthly_return_per_ticker.csv")
monthly_df.to_sql("stock_monthly_metrics", engine, if_exists="replace", index=False)
```
