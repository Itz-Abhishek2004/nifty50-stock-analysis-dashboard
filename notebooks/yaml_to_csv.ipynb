```python
import os
import yaml
import pprint as pp

yaml_root = "../data_raw/data"  # yaha pe 'data' tumhara actual main folder

# Step 1: month folders ki list
months = [d for d in os.listdir(yaml_root) 
          if os.path.isdir(os.path.join(yaml_root, d))]
print("Month folders:", months)

# Step 2: kisi ek month ke andar ki YAML files
sample_month = months[0]
month_path = os.path.join(yaml_root, sample_month)
yaml_files = [f for f in os.listdir(month_path) 
              if f.endswith(".yaml") or f.endswith(".yml")]
print("Sample month:", sample_month)
print("YAML files (first 5):", yaml_files[:5])

# Step 3: ek YAML file load karo
sample_file = yaml_files[0]
file_path = os.path.join(month_path, sample_file)
print("Using file:", file_path)

with open(file_path, "r") as f:
    data = yaml.safe_load(f)

# structure dekhne ke liye
if isinstance(data, dict):
    pp.pprint(list(data.items())[:5])
else:
    print(type(data))
    pp.pprint(data)
```

```python
import os
import yaml
import pandas as pd

yaml_root = "../data_raw/data"   # yaha pe apne main YAML folder ka naam

all_rows = []

months = [d for d in os.listdir(yaml_root) 
          if os.path.isdir(os.path.join(yaml_root, d))]

print("Months found:", months)

for month in months:
    month_path = os.path.join(yaml_root, month)
    yaml_files = [f for f in os.listdir(month_path) 
                  if f.endswith(".yaml") or f.endswith(".yml")]
    
    print(f"Processing month: {month}, files: {len(yaml_files)}")
    
    for yf in yaml_files:
        file_path = os.path.join(month_path, yf)
        
        with open(file_path, "r") as f:
            data = yaml.safe_load(f)   # list of dicts
        
        if not isinstance(data, list):
            continue
        
        for row in data:
            all_rows.append(row)

len(all_rows)
```

```python
df = pd.DataFrame(all_rows)
df.head(), df.shape, df.columns
```

```python
# basic cleaning
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values(['Ticker', 'date'])

# rename columns for consistency
df = df.rename(columns={
    'Ticker': 'ticker',
    'open': 'open',
    'high': 'high',
    'low': 'low',
    'close': 'close',
    'volume': 'volume',
    'month': 'month'
})

df.head()
```
